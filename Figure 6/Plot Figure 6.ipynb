```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
### Ian's code:

# Define the file path
file_path = r'..\Data\Final table for M_surv by M_ori for M greater than 1e-12 Msolar.xlsx' # ~/Dropbox/Ian/Chris/Finaltable1.xlsx'

# Load the data from the sheet
df = pd.read_excel(file_path, sheet_name='Sheet1')

# Define the columns based on the data
x_data = df['Axion mass (micro eV)']
y_data = df['M_surv/M_ori (>1e-12  Msolar) [%]']
m_min_data = df['M_min [Msolar]']
p_data = df['p']

# Plotting
# plt.figure(figsize=(10, 6))
plt.rcParams["mathtext.fontset"] = "cm"

for i in range(int(x_data.size / 3)): # Divide by 3 because the rows are bunched by 3 axion masses
    if p_data[i*3] == 1:
        color = 'b'
    elif p_data[i*3] == 2:
        color = 'r'
    else:
        color = 'g'  # Just in case there are other values of p
    
    if m_min_data[i*3] == 0.01:
        marker = 'x'
        markeredgewidth = 2
    elif m_min_data[i*3] == 100:
        marker = 'd'
        markeredgewidth = 1

    plt.plot(x_data[i*3 : (i+1)*3], y_data[i*3 : (i+1)*3], marker=marker, markeredgewidth=markeredgewidth, color=color, linestyle="dashed")

# Creating legend info
plt.plot([], [], color="blue", linestyle="dashed", label=r"$p=\text{hybrid}$")
plt.plot([], [], color="red", linestyle="dashed", label=r"$p=2$")
plt.scatter([], [], marker="x", linewidths=2 , color="black", label=r"$M_{\rm min} = 10^{-2}M_\odot$")
plt.scatter([], [], marker="d", color="black", label=r"$M_{\rm min} = 10^{2}M_\odot$")

plt.xscale('log')
plt.xlabel(r"$m_{\rm a} [\mu \text{eV}]$", fontsize=18)
plt.ylabel(r"$M_{\rm surv}/M_{\rm ori}\ (M>10^{-12} M_\odot) [\%]$", fontsize=18)
# plt.title('Plot of M_surv/M_ori vs. Axion mass (Log Scale)')
plt.xlim(1e0, 1e3)
plt.grid(True)
plt.legend(fontsize=13, loc=(0.7, 0.27))
plt.tight_layout()
# plt.savefig("MSurv_By_MOri_for_M_greater_than_1e-12.pdf")
plt.show()
```
