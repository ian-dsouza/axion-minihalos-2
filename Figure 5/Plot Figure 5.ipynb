```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load all files - bin file and M_surv/M_ori files

bins = np.load(r"..\Data\bins.npy")

# File 11 - p=2; M_min = 1e-2 MSolar
mSurv_by_mOri_file11_ma_1p25 = np.load(r"..\Data\File 11\M_surv by M_ori - 1.25mueV.npy")
mSurv_by_mOri_file11_ma_25 = np.load(r"..\Data\File 11\M_surv by M_ori - 25mueV.npy")
mSurv_by_mOri_file11_ma_500 = np.load(r"..\Data\File 11\M_surv by M_ori - 500mueV.npy")

# File 12 - p=hybrid; M_min = 1e-2 MSolar
mSurv_by_mOri_file12_ma_1p25 = np.load(r"..\Data\File 12\M_surv by M_ori - 1.25mueV.npy")
mSurv_by_mOri_file12_ma_25 = np.load(r"..\Data\File 12\M_surv by M_ori - 25mueV.npy")
mSurv_by_mOri_file12_ma_500 = np.load(r"..\Data\File 12\M_surv by M_ori - 500mueV.npy")

# File 15 - p=2; M_min = 1e2 MSolar
mSurv_by_mOri_file15_ma_1p25 = np.load(r"..\Data\File 15\M_surv by M_ori - 1.25mueV.npy")
mSurv_by_mOri_file15_ma_25 = np.load(r"..\Data\File 15\M_surv by M_ori - 25mueV.npy")
mSurv_by_mOri_file15_ma_500 = np.load(r"..\Data\File 15\M_surv by M_ori - 500mueV.npy")

# File 16 - p=hybrid; M_min = 1e2 MSolar
mSurv_by_mOri_file16_ma_1p25 = np.load(r"..\Data\File 16\M_surv by M_ori - 1.25mueV.npy")
mSurv_by_mOri_file16_ma_25 = np.load(r"..\Data\File 16\M_surv by M_ori - 25mueV.npy")
mSurv_by_mOri_file16_ma_500 = np.load(r"..\Data\File 16\M_surv by M_ori - 500mueV.npy")
```

Graph for M_min = 1e-2 MSolar

```python
plt.rcParams["mathtext.fontset"] = "cm"

plt.semilogx(bins, mSurv_by_mOri_file11_ma_1p25, color="red", linestyle="dashed")
plt.semilogx(bins, mSurv_by_mOri_file11_ma_25, color="blue", linestyle="dashed")
plt.semilogx(bins, mSurv_by_mOri_file11_ma_500, color="green", linestyle="dashed")

plt.semilogx(bins, mSurv_by_mOri_file12_ma_1p25, color="red")
plt.semilogx(bins, mSurv_by_mOri_file12_ma_25, color="blue")
plt.semilogx(bins, mSurv_by_mOri_file12_ma_500, color="green")

plt.scatter([], [], color="red", label=r"$m_{\rm a} = 1.25 \mu\text{eV}$")
plt.scatter([], [], color="blue", label=r"$m_{\rm a} = 25 \mu\text{eV}$")
plt.scatter([], [], color="green", label=r"$m_{\rm a} = 500 \mu\text{eV}$")
plt.semilogx([], [], color="black", linestyle="dashed", label=r"$p=2$")
plt.semilogx([], [], color="black", label=r"$p=\text{hybrid}$")


plt.xlabel(r"$M [M_\odot]$", fontsize=18)
plt.ylabel(r"$M_{\rm surv.} / M_{\rm ori.} (>M)$", fontsize=18)
plt.ylim(0, 1)
plt.xlim(1e-12, 1e-5)
plt.legend(fontsize=14)
plt.tight_layout()
# plt.savefig("MSurv_by_MOri_Mmin=1e-2MSolar.pdf")
plt.show()
```
